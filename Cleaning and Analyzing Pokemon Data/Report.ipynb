General Thoughts what Pokemon Type has the best advantage against the other types. And What Pokemon specifically (non Legendary and legendary) has the best advantage against the others. 

```python
dataframe.columns

new_order = ['pokedex_number', 'name','type1', 'type2', 'hp', 'attack', 'defense',
'sp_defense', 'sp_attack',  'speed', 'abilities', 'against_bug', 'against_dark', 'against_dragon',
'against_electric', 'against_fairy', 'against_fight', 'against_fire',
'against_flying', 'against_ghost', 'against_grass', 'against_ground',
'against_ice', 'against_normal', 'against_poison', 'against_psychic',
'against_rock', 'against_steel', 'against_water',
'base_egg_steps', 'base_happiness', 'base_total', 'capture_rate',
'classfication', 'experience_growth', 'height_m', 
'japanese_name', 'percentage_male','weight_kg','generation', 'is_legendary']

df = dataframe.reindex(columns = new_order)

#To Accomidate if there are pokemon with no Secondary Typing
df['type2'] = df['type2'].fillna("None")
df['Merged_Type'] = df['type1'] + "-" + df['type2']
grouped_merged = df.groupby('Merged_Type').size().reset_index(name='Count')
grouped_merged

grouped_type1 = df.groupby('type1').size().reset_index(name='Count')
grouped_type2 = df.groupby('type2').size().reset_index(name='Count')
```

```python
# fig, axes = plt.subplots(2,3, figsize=(10,6), constrained_layout = True, sharey=True)
# fig.suptitle("Pokemon Stats - Typing")

# sns.boxplot(ax=axes[0,0], data=df, x ='type1', y='hp', native_scale=True).tick_params(axis='x',rotation =90).set(title="HP")
# sns.boxplot(ax=axes[0,1], data=df, x ='type1', y='attack', native_scale=True).tick_params(axis='x',rotation =90).set(title="Attack")
# sns.boxplot(ax=axes[0,2], data=df, x ='type1', y='defense', native_scale=True).tick_params(axis='x',rotation =90).set(title="Defense")
# sns.boxplot(ax=axes[1,0], data=df, x ='type1', y='sp_attack', native_scale=True).tick_params(axis='x',rotation =90).set(title="Special Attack")
# sns.boxplot(ax=axes[1,1], data=df, x ='type1', y='sp_defense', native_scale=True).tick_params(axis='x',rotation =90).set(title="Special Defense")
# sns.boxplot(ax=axes[1,2], data=df, x ='type1', y='speed', native_scale=True).tick_params(axis='x',rotation =90).set(title="Speed")



fig, axes = plt.subplots(2,3, figsize=(10,6), constrained_layout = True, sharey=True)
fig.suptitle("Pokemon Stats - Typing")

fig1 = sns.boxplot(ax=axes[0,0], data=df, x ='type1', y='hp', native_scale=True)
fig1.set(title="Health")
fig1.tick_params(axis='x',rotation =90)

fig2 = sns.boxplot(ax=axes[0,1], data=df, x ='type1', y='attack', native_scale=True)
fig2.set(title="Attack")
fig2.tick_params(axis='x',rotation =90)

fig3 = sns.boxplot(ax=axes[0,2], data=df, x ='type1', y='defense', native_scale=True)
fig3.set(title="Defense")
fig3.tick_params(axis='x',rotation =90)

fig4 = sns.boxplot(ax=axes[1,0], data=df, x ='type1', y='sp_attack', native_scale=True)
fig4.set(title='Special Attack')
fig4.tick_params(axis='x',rotation =90)

fig5 = sns.boxplot(ax=axes[1,1], data=df, x ='type1', y='sp_defense', native_scale=True)
fig5.set(title="Special Defense")
fig5.tick_params(axis='x',rotation =90)

fig6 = sns.boxplot(ax=axes[1,2], data=df, x ='type1', y='speed', native_scale=True)
fig6.set(title="Speed")
fig6.tick_params(axis='x',rotation =90)
```

```python
All_columns = ['against_bug', 'against_dark', 'against_dragon',
'against_electric', 'against_fairy', 'against_fight', 'against_fire',
'against_flying', 'against_ghost', 'against_grass', 'against_ground',
'against_ice', 'against_normal', 'against_poison', 'against_psychic',
'against_rock', 'against_steel', 'against_water']
df['Total_Against'] = df[All_columns].sum(axis=1)


New_Columns_For_Dataframe = ["Total_Against", "type1", "type2", "Merged_Type", 'hp', 'attack', 'defense',
'sp_defense', 'sp_attack',  'speed']


new_dataframe = df[New_Columns_For_Dataframe]

fig1 = sns.boxplot(data=new_dataframe, x="type1", y="Total_Against", native_scale=True)
fig1.tick_params(axis='x',rotation =90)


```

```python
new_dataframe


average_of_group = df.groupby('type1')['Total_Against'].mean().sort_values(ascending=True).reset_index(name="Count")


fig1 = sns.barplot(data = average_of_group, x = "type1", y = "Count")
fig1.tick_params(axis='x',rotation =90)
fig1.set(title="Total Against Specific Typing")
```
